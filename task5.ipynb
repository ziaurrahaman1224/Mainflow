```python
import tkinter as tk
from tkinter import ttk
from datetime import datetime
import requests
from PIL import ImageTk, Image
from tkinter import messagebox

# Create the main window
root = tk.Tk()
root.geometry("600x270")
root.title("Currency Converter")
root.iconbitmap("C:\\Users\\ziaur\\Downloads\\icon.ico")
root.maxsize(600, 270)
root.minsize(600, 270)

# Load and display the image
image = Image.open("C:\\Users\\ziaur\\Downloads\\currency.png")
zoom = 0.5
pixels_x, pixels_y = tuple(int(zoom * i) for i in image.size)
image = ImageTk.PhotoImage(image.resize((pixels_x, pixels_y)))
panel = tk.Label(root, image=image)
panel.place(x=190, y=35)

# Function to show data
def show_data():
    amount = E1.get()
    from_currency = c1.get()
    to_currency = c2.get()

    if not amount:
        messagebox.showerror("Currency Converter", "Please fill in the amount")
        return

    if not to_currency:
        messagebox.showerror("Currency Converter", "Please choose the currency")
        return

    try:
        amount = float(amount)
    except ValueError:
        messagebox.showerror("Currency Converter", "Invalid amount")
        return

    try:
        url = f"https://api.exchangerate-api.com/v4/latest/{from_currency}"
        data = requests.get(url).json()
        
        if 'error' in data:
            messagebox.showerror("Currency Converter", data['error'])
            return
        
        exchange_rate = data['rates'][to_currency]
        converted_amount = amount * exchange_rate
        E2.delete(0, tk.END)
        E2.insert(0, f"{converted_amount:.2f}")
        text.insert(tk.END, f"{amount} {from_currency} equals {converted_amount:.2f} {to_currency}\n")
        text.insert(tk.END, f"Last Updated: {datetime.now()}\n")
    except Exception as e:
        messagebox.showerror("Currency Converter", str(e))

# Function to clear data
def clear():
    E1.delete(0, tk.END)
    E2.delete(0, tk.END)
    text.delete(1.0, tk.END)

# Labels and Entries
label_title = tk.Label(root, text="USD Currency Converter Using Python", font=('verdana', 10, 'bold'))
label_title.place(x=150, y=15)

label_amount = tk.Label(root, text="Amount", font=('roboto', 10, 'bold'))
label_amount.place(x=20, y=15)
E1 = tk.Entry(root, width=20, borderwidth=1, font=('roboto', 10, 'bold'))
E1.place(x=20, y=40)

label_from_currency = tk.Label(root, text="From Currency", font=('roboto', 10, 'bold'))
label_from_currency.place(x=20, y=70)
c1 = tk.StringVar()
currencychoosel = ttk.Combobox(root, width=20, textvariable=c1, state='readonly', font=('verdana', 10, 'bold'))
currencychoosel['values'] = ("USD", "EUR", "GBP", "INR", "AUD", "CAD", "SGD", "JPY", "CNY", 
                             "CHF", "HKD", "NZD", "ZAR", "RUB", "BRL", "MXN", "SEK", "NOK", 
                             "DKK", "PLN", "THB", "IDR", "CZK", "HUF", "ILS", "MYR", "KRW", 
                             "PHP", "AED", "SAR", "TRY", "EGP", "PKR", "BDT", "VND", "COP",
                             "ARS", "CLP", "PEN", "LKR", "NPR", "MAD", "TWD", "KES", "UGX", 
                             "GHS", "XOF", "TND", "UAH", "QAR", "KWD", "OMR", "BHD", "IQD")
currencychoosel.place(x=20, y=90)
currencychoosel.current(0)

label_to_currency = tk.Label(root, text="To Currency", font=('roboto', 10, 'bold'))
label_to_currency.place(x=300, y=15)
c2 = tk.StringVar()
currencychoose2 = ttk.Combobox(root, width=20, textvariable=c2, state='readonly', font=('verdana', 10, 'bold'))
currencychoose2['values'] = ("USD", "EUR", "GBP", "INR", "AUD", "CAD", "SGD", "JPY", "CNY", 
                             "CHF", "HKD", "NZD", "ZAR", "RUB", "BRL", "MXN", "SEK", "NOK", 
                             "DKK", "PLN", "THB", "IDR", "CZK", "HUF", "ILS", "MYR", "KRW", 
                             "PHP", "AED", "SAR", "TRY", "EGP", "PKR", "BDT", "VND", "COP",
                             "ARS", "CLP", "PEN", "LKR", "NPR", "MAD", "TWD", "KES", "UGX", 
                             "GHS", "XOF", "TND", "UAH", "QAR", "KWD", "OMR", "BHD", "IQD")
currencychoose2.place(x=300, y=40)
currencychoose2.current(1)

E2 = tk.Entry(root, width=20, borderwidth=1, font=('roboto', 10, 'bold'))
E2.place(x=300, y=90)

text = tk.Text(root, height=7, width=52, font=('verdana', 10, 'bold'))
text.place(x=100, y=120)

# Buttons
B = tk.Button(root, text="Convert", command=show_data, font=('verdana', 10, 'bold'), borderwidth=2, bg="red", fg="white")
B.place(x=20, y=120)

clear_button = tk.Button(root, text="Clear", command=clear, font=('verdana', 10, 'bold'), borderwidth=2, bg="blue", fg="white")
clear_button.place(x=20, y=170)

root.mainloop()
```
